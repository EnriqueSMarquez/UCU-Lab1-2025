```python
import matplotlib
import matplotlib.pyplot as plt
import json

with open('./run4-test/training_metrics.json', 'r') as f:
    metrics = json.load(f)
```

```python
import numpy as np
epochs = np.arange(0, 30, 1)

plt.figure(figsize=(12, 8))

plt.plot(epochs, metrics['train_loss'], label='training_loss')
plt.plot(epochs, metrics['train_acc'], label='training_acc')
plt.plot(epochs, metrics['val_loss'], label='validation_loss')
plt.plot(epochs, metrics['val_acc'], label='validation_acc')

plt.xlabel('Epochs')
plt.ylabel('Loss / Acc')

plt.legend(loc=0)
```
